```python
# Import Required Dependencies
import pandas as pd
import requests
from bs4 import BeautifulSoup

# base url of OLX website
base_url = 'https://www.olx.com.pk'

# give the url which product data we want to retrive
url = 'https://www.olx.com.pk/items/q-honda-city'

# send the request
html_parse = requests.get(url)

soup = BeautifulSoup(html_parse.content, 'lxml')

print('Process has started!!!')
# make a list for Products links
product_links = []

# find all 'li' tag in the class 
product_list = soup.find_all('li', class_ = 'c46f3bfe')

# apply loop on the list to tackle with each product
for item in product_list:
    # Find all the 'a' tags to retrieve the links 
    for i in item.find_all('a'):
        # concatenate the find link with the base link
        link = base_url + i.attrs['href']
        # apply condition if the link is not in the list then add it into the list
        if link not in product_links:
            product_links.append(link)

print('Number of Products Links: ',len(product_links))

#Header Information
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

# make a dictionary and give the keys of which data do we want
products = {'titles' : [],
            'prices' : [],
            'models' : [],
            'mileages' : [],
            'locations' : [],
            'posted date' : []}

# stores dictionary keys into keys variable
keys = list(products.keys())

for link in product_links:
    # send request to the link
    r = requests.get(link, headers = header)
    soup = BeautifulSoup(r.content, 'lxml')
    
    product_information = []
    
    product_information.append(soup.find('h1').text)
    
    # apply a condition on 'span' tag which has the information of price. sometimes prices are not given on the ad
    try:
        product_information.append(soup.find('span', class_ = '_56dab877').text)
    except:
        product_information.append('no price')
        
    for a in soup.find('span', class_ = '_261203a9 _7f02578b').find('span', class_ = 'fef55ec1').text.split('â€¢'):
        product_information.append(a)
    
    location_posted = []
    
    for i in soup.find('div', class_ ="_1075545d e3cecb8b _5f872d11").find_all('span', class_ = '_8918c0a8'):
        product_information.append(i.text)
    
    # append all the necessary information in the dictionary using keys vairable
    for index, i in enumerate(keys):
        products[i].append(product_information[index])

print('Task Complete!!!!')

# convert into Pandas DataFrame
product_df = pd.DataFrame(products)

file_name = input('Enter the File Name: ')
# save in CSV format in local machine
product_df.to_csv(file_name + '.csv', index = False)
print('CSV file has saved in your local machine')
```
