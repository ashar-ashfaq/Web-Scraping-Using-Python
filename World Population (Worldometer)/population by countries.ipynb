### Link : https://www.worldometers.info/world-population/population-by-country/

```python
# Import Required Dependencies
import requests
from bs4 import BeautifulSoup
import pandas as pd
# Page Url
url = 'https://www.worldometers.info/world-population/population-by-country/'

# send the request
html_text = requests.get(url)

# scraping the whole content of the page
soup = BeautifulSoup(html_text.content, 'html.parser')

titles = []

# scraping the table headers
headings = soup.find('thead')
for title in headings.tr.find_all('th'):
    titles.append(title.text)

data = dict()

for i in titles:
    if i not in data.keys():
        data[i] = []

keys = list(data.keys())
for i in soup.find('tbody').find_all('tr'):
    for index,j in enumerate(i.find_all('td')):
        data[keys[index]].append(j.text)

# Loading the data into pandas dataframe
population_by_counrty = pd.DataFrame(data)
population_by_counrty.drop('#', axis=1, inplace = True)

# saving the Dataframe to csv file 
population_by_counrty.to_csv('population by world.csv', index = False)
```
